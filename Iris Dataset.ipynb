### Applying Machine Learning

```python
# Train_test split

from sklearn import model_selection
irisarr = iris.values
X = irisarr[:,0:4]
Y = irisarr[:,4]
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = \
    model_selection.train_test_split(X,Y, test_size = validation_size , random_state=seed)
```

### Summary 

Accuracy -  (true positive + true negative) / total predictions

metrics.precision_score(true, pred)

Precision  - true positive / (true positive + false positive)

metrics.precision_score(true, pred)

Recall - (sensitivity) true positive / (true positive + false negative)

metrics.recall_score(true, pred) F1-Score 2 * (precision * recall) / (precision + recall)

metrics.f1_score(true, pred)

Specificity - true negative / (true negative + false positive)

metrics.recall_score(true, pred, pos_label=0)

### Visualizing accuracy score, confusion matrix through heatmap- Support Vector Classifier(SVC)

SVC works by mapping data points to a high-dimensional space and then finding the optimal hyperplane that divides the data into two classes.

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns  

clf = SVC(kernel='linear').fit(X_train, Y_train)
y_pred = clf.predict(X_validation)

cm = confusion_matrix(Y_validation, y_pred)

cm_df = pd.DataFrame(cm,
                    index=['setosa', 'versicolor', 'virginica'],
                    columns=['setosa', 'versicolor', 'virginica'])

plt.figure(figsize=(5.5, 4))
sns.heatmap(cm_df, annot=True)

# Calculate accuracy
accuracy = accuracy_score(Y_validation, y_pred)

# Corrected title formatting
plt.title('SVM Linear Kernel \nAccuracy: {:.3f}'.format(accuracy))
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

### Visualizing Accuracy score, confusion matrix through heatmap-Linear Discriminant Analysis (LDA)

Linear discriminant analysis (LDA), normal discriminant analysis (NDA), or discriminant function analysis is a generalization of Fisher's linear discriminant, a method used in statistics and other fields, to find a linear combination of features that characterizes or separates two or more classes of objects or events.

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Assuming seaborn is installed
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# ... (Assuming you have your data loaded)

# Split data into training and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create the LinearDiscriminantAnalysis model
clf = LinearDiscriminantAnalysis()
clf.fit(X_train, Y_train)

# Make predictions on the validation set
y_pred = clf.predict(X_val)

# Calculate the confusion matrix
cm = confusion_matrix(Y_val, y_pred)

# Create the confusion matrix DataFrame
cm_df = pd.DataFrame(cm,
                    index=['setosa', 'versicolor', 'virginica'],
                    columns=['setosa', 'versicolor', 'virginica'])

# Create the heatmap
plt.figure(figsize=(5.5, 4))
sns.heatmap(cm_df, annot=True)

# Calculate accuracy
accuracy = accuracy_score(Y_val, y_pred)

# Corrected title formatting (avoid extra backslashes)
plt.title('LinearDiscriminantAnalysis\nAccuracy: {:.3f}'.format(accuracy))
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

### Visualizing accuracy score, confusion matrix through heatmap-GaussianNB

Gaussian Naive Bayes (GNB) is a classification technique used in Machine Learning (ML) based on the probabilistic approach and Gaussian distribution. Gaussian Naive Bayes assumes that each parameter (also called features or predictors) has an independent capacity of predicting the output variable.

Gaussian Naïve Bayes is used when we assume all the continuous variables associated with each feature to be distributed according to Gaussian Distribution. Gaussian Distribution is also called Normal distribution. We can use this formula to compute the probability of likelihoods if our data is continuous.

Naive Bayes is a supervised classification algorithm that is used primarily for dealing with binary and multi-class classification problems, though with some modifications, it can also be used for solving regression problems.

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB().fit(X_train,Y_train)
clf.predict(X_train)
y_pred = clf.predict(X_validation)

# Creates a confusion matrix
cm = confusion_matrix(Y_validation, y_pred)

# Transform to df for easier plotting
cm_df = pd.DataFrame(cm,
                     index = ['setosa','versicolor','virginica'],
                     columns = ['setosa','versicolor','virginica'])

plt.figure(figsize=(5.5,4))
sns.heatmap(cm_df, annot=True)
plt.title('GaussianNB \nAccuracy:{0:.3f}'.format(accuracy_score(Y_validation, y_pred)))
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```
